## Question A: Posterior of the Speed of Each Object 

Object A is **blue**. 

Object B is **red**.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm
from scipy import stats 

%matplotlib inline

datt = pd.read_csv('location_data_hw9.csv')
datt.head()
````

````python
datt.describe()
````

````python
rand_dir = np.random.choice([-1, 1], 1)[0]
rand_dir
````

````python
with pm.Model() as modd:
    # Setup prior distributions for positions
    
    # Get 4 samples from random uniform distributions (A_x, A_y, B_x, B_y) 
    vel = pm.Uniform('vel', lower = -1.5, upper = 1.5, shape = 4)
    sigg = pm.HalfNormal('sigg', sd = 0.5, shape = 4)
    init_pos_x = pm.Uniform('init_pos_x', lower = 0.9, upper = 1.6, shape = 2)
    init_pos_y = pm.Uniform('init_pos_y', lower = -1.5, upper = 1.5, shape = 2)
    
    # Generate random data from real summary: step 1 of 2 
    mu_pos_Ax = vel[3] * datt['t'] + init_pos_x[1]
    mu_pos_Ay = vel[2] * datt['t'] + init_pos_y[1]
    mu_pos_Bx = vel[1] * datt['t'] + init_pos_x[0]
    mu_pos_By = vel[0] * datt['t'] + init_pos_y[0]
    
    # Generate random data from real summary: step 2 of 2 
    nu_Ax = pm.Normal('blue_nu_x', mu = mu_pos_Ax, sd = sigg[0], observed = datt['blue_pos_X'])
    nu_Ay = pm.Normal('blue_nu_y', mu = mu_pos_Ay, sd = sigg[1], observed = datt['blue_pos_Y'])
    nu_Bx = pm.Normal('red_nu_x', mu = mu_pos_Bx, sd = sigg[2], observed = datt['red_pos_X'])
    nu_By = pm.Normal('red_nu_y', mu = mu_pos_By, sd = sigg[3], observed = datt['red_pos_Y'])
````

````python
with modd:
    step = pm.Metropolis()
    trace = pm.sample(10000, step = step, njobs = 1)
````

````python
pos_inn = trace[5000:]
pos_thinn = pos_inn[::10]
pm.traceplot(pos_thinn)
````

**Plots above and summary statistics (table) below address part a) and used for b) to d)**

````python
## A_x (blue) = vel_3, sigg_0 
## A_y (blue) = vel_2, sigg_1
## B_x (red) = vel_1, sigg_2
## B_y (red) = vel_0, sigg_0

model_summ = pm.df_summary(pos_thinn, alpha = 0.1)
````

The values for the model (above) are based off of the summary / description of the real data. 

Our equations for speed and position: 

*new_position_x = speed_x * time + initial_x* 

So - with rearranging some terms we can get: 

```
    pos_y = vel_y * time + init_y 
    
    pos_y - init_y = vel_y * time 
    
    time = pos_y - init_y / vel_y 
```

With **pos_y = 0**, we get: 

```
    time = -init_y / vel_y 
```

The posterior of the speed of each object can be found by looking at the summary or statistical descriptions of the simulated velocity, initial positions, and new positions in respect to time.

## Question C: Same Speed? 

To create equal magnitude of velocities:

```
    1. Generate uniform random magnitude = sum_vel_sq
    2. Generate uniform random on [0, sum_vel_sq] = A_vel_x_sq  
    3. Repeat step 2 = B_vel_x_sq 
    4. Find difference between sum_vel_sq and A_vel_x_sq = A_vel_y_sq 
    5. Repeat step 4 but for B = B_vel_y_sq
    6. Take square root to generate = A_vel_x, A_vel_y, B_vel_x, B_vel_y 
    7. Apply Bernoulli (i.e. -1 or 1) as directionality scaler  
```

````python
with pm.Model() as modd_c:
    # Setup prior distributions for positions
    
    # Get equal magnitudes of velocity (A_x, A_y, B_x, B_y) 
    # sum_vel_sq = v_x ** 2 + v_y ** 2
    # A_vel_x_sq = pm.Uniform(0, sum_vel_sq) -> Random on [0, sum_vel_sq]
    # A_vel_y_sq = sum_vel_sq - A_vel_x_sq
    # A_vel_x = square root of (A_vel_x_sq)
    # A_vel_y = square root of (A_vel_y_sq) 
    
    sum_vel_sq = pm.Uniform('sum_vel_sq', lower = -2, upper = 2, shape = 1)
    A_vel_x_sq = pm.Uniform('A_vel_x_sq', lower = 0, upper = sum_vel_sq, shape = 1)
    B_vel_x_sq = pm.Uniform('B_vel_x_sq', lower = 0, upper = sum_vel_sq, shape = 1)
    
    #sum_vel_sq = np.random.uniform(low = -2, high = 2, size = 1)
    #A_vel_x_sq = np.random.uniform(low = 0, high = sum_vel_sq, size = 1)
    #B_vel_x_sq = np.random.uniform(low = 0, high = sum_vel_sq, size = 1)
    A_vel_y_sq = sum_vel_sq[0] - A_vel_x_sq[0]
    B_vel_y_sq = sum_vel_sq[0] - B_vel_x_sq[0]
    
    ## (A_vel_x_sq + A_vel_y_sq) == (B_vel_x_sq + B_vel_y_sq)
    #A_vel_x = np.sqrt(A_vel_x_sq)
    #A_vel_y = np.sqrt(A_vel_y_sq)
    #B_vel_x = np.sqrt(B_vel_x_sq)
    #B_vel_y = np.sqrt(B_vel_y_sq)
    
    sigg = pm.HalfNormal('sigg', sd = 0.5, shape = 4)
    init_pos_x = pm.Uniform('init_pos_x', lower = 0.9, upper = 1.6, shape = 2)
    init_pos_y = pm.Uniform('init_pos_y', lower = -1.5, upper = 1.5, shape = 2)
    
    # Random directionality (could be Bernoulli as well)
    check_switch = pm.Uniform('check_switch', lower = 0, upper = 2, shape = 4)
    rand_dir1 = pm.math.switch(check_switch[0] >= 1, -1, 1)
    rand_dir2 = pm.math.switch(check_switch[1] >= 1, -1, 1)
    rand_dir3 = pm.math.switch(check_switch[2] >= 1, -1, 1)
    rand_dir4 = pm.math.switch(check_switch[3] >= 1, -1, 1)
    
    #rand_dir1 = np.random.choice([-1, 1], 1)[0]
    #rand_dir2 = np.random.choice([-1, 1], 1)[0]
    #rand_dir3 = np.random.choice([-1, 1], 1)[0]
    #rand_dir4 = np.random.choice([-1, 1], 1)[0]
    
    # Generate random data from real summary: step 1 of 2 
    mu_pos_Ax = A_vel_x_sq * rand_dir1 * datt['t'] + init_pos_x[0]
    mu_pos_Ay = A_vel_y_sq * rand_dir2 * datt['t'] + init_pos_y[0]
    mu_pos_Bx = B_vel_x_sq * rand_dir3 * datt['t'] + init_pos_x[1]
    mu_pos_By = B_vel_y_sq * rand_dir4 * datt['t'] + init_pos_y[1]

    # Generate random data from real summary: step 2 of 2 
    nu_Ax = pm.Normal('blue_nu_x', mu = mu_pos_Ax, sd = sigg[0], observed = datt['blue_pos_X'])
    nu_Ay = pm.Normal('blue_nu_y', mu = mu_pos_Ay, sd = sigg[1], observed = datt['blue_pos_Y'])
    nu_Bx = pm.Normal('red_nu_x', mu = mu_pos_Bx, sd = sigg[2], observed = datt['red_pos_X'])
    nu_By = pm.Normal('red_nu_y', mu = mu_pos_By, sd = sigg[3], observed = datt['red_pos_Y'])
````

````python
with modd_c:
    step_c = pm.Metropolis()
    trace_c = pm.sample(10000, step = step_c, njobs = 1)
    
burn_c = trace_c[5000:]
trace_thin_c = burn_c[::10]
pm.traceplot(trace_thin_c)
````
